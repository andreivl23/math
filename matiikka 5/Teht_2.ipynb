# Funktioiden kuvaajien piirtäminen Pythonilla

Python on hyvä työkalu funktioiden kuvaajien piirtämisen. Tarvitaan kirjastot `matplotlib`, sieltä varsinkin `pyplot`, isäksi `numpy`. Eli alla olevat määritykset pitää olla:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Tehtävä 1

```python
x = np.linspace(-3, -1, 100)
x1 = np.linspace(-5, -3, 100)
x2 = np.linspace(-1, 1, 100)
x3 = np.linspace(3, 5, 100)
x4 = np.linspace(1, 3, 100)
x5 = np.linspace(5, 7, 100)
y1 = 0*x+4

z = -2*x-2

ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')

# x-akselin säätöä
ax.set_xticks(np.arange(-5, 7, 1))
ax.set_yticks(np.arange(0, 5, 1))


# Y:lle rajat, että kuvaajat näkyvät fiksummin
plt.ylim([0, 4])
plt.xlim([-5,7])

# label tarvitsee tuon legend():n
plt.plot(x1,y1, "blue")
plt.plot(x2,y1, "blue")
plt.plot(x3,y1, "blue")

plt.plot(x, z, "blue")
plt.plot(x4,z, "blue")
plt.plot(x5,z, "blue")

plt.show()
```

# Tehtävä 2

```python
x = np.linspace(-5, 0, 100)
x1 = np.linspace(0, 4, 100)
x2 = np.linspace(4, 8, 100)
x3 = np.linspace(8, 13, 100)

y1 = 0*x+3
y2 = (3/5)*x+3
y3 = -(3/5)*x
ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')

# x-akselin säätöä
ax.set_xticks(np.arange(-5, 11, 1))
ax.set_yticks(np.arange(0, 5, 1))

# Y:lle rajat, että kuvaajat näkyvät fiksummin
plt.ylim([0, 4])
plt.xlim([-5,10])

# label tarvitsee tuon legend():n
plt.plot(x,y2, "blue")
plt.plot(x1,y1, "blue")
plt.plot(x2,y3, "blue")
plt.plot(x3,y2, "blue")


plt.show()
```

# Tehtävä 3

```python
x = np.linspace(-5, 5, 100)
y = x**2-4*x+3
z = -1.5*x**2-3*x+7

ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')

# x-akselin säätöä
ax.set_xticks(np.arange(-5, 5, 1))


# Y:lle rajat, että kuvaajat näkyvät fiksummin
plt.ylim([-9, 9])

# label tarvitsee tuon legend():n
plt.plot(x, y, label="$y= x^2−4 x+ 3$")
plt.plot(x, z, label="$y=−1,5 x^2−3 x+ 7$")
plt.legend()

plt.show()
```

# Tehtävä 4

```python
x = np.linspace(0, 150, 10)
y = -0.0063*x**2+0.55*x


ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')

# x-akselin säätöä
ax.set_xticks(np.arange(0, 100, 10))


# Y:lle rajat, että kuvaajat näkyvät fiksummin
plt.ylim([0, 20])

# label tarvitsee tuon legend():n
plt.plot(x, y, label="$y =−0,0063 x^2+ 0,55 x$")

plt.legend()

plt.show()
```
